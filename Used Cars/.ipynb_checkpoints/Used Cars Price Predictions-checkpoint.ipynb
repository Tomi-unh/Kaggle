```python
import re

#get a column for the horsepower and the engine volume 
def extract_volume_horsepower(description):
    volume_match = re.search(r'(\d+(\.\d+)?)L', description)
    hp_match = re.search(r'(\d+(\.\d+)?)hp', description, re.IGNORECASE)
    
    volume = float(volume_match.group(1)) if volume_match else None
    horsepower = float(hp_match.group(1)) if hp_match else None
    
    return volume, horsepower

train_df[['volume', 'horsepower']] = train_df['engine'].apply(lambda x: pd.Series(extract_volume_horsepower(x)))
```

## Encoding using an LLM 

```python
#combine the string columns and parse into the LLM 
cat_df['combined'] = cat_df.apply(lambda x:' '.join(x), axis = 1)
```

```python
cat_df.head()
```

```python
#build LLM model
from transformers import BertTokenizer, BertModel

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

def encode_combined(combined):
    inputs = tokenizer(combined, return_tensors='pt')
    outputs = model(**inputs)
    # Process the outputs to get the mean-pooled embedding
    embedding = outputs.last_hidden_state.mean(dim=1).squeeze().detach().numpy()
    return embedding

cat_df['encode'] = cat_df['combined'].apply(encode_combined)
```

```python
#save encoded dataframe
cat_df.to_pickle('categories_with_embeddings.pkl')
```

```python
#fill the horsepower and volume column with the average of the cars based on brand
#if there is only one car for the brand, fill with average of all the horsepower 

#add brand to the df 
train_df = pd.read_csv('train.csv')

train_df.drop(['id', 'clean_title'], axis = 1, inplace = True)

llm_df['brand'] = train_df['brand']
```

```python
def fill_nan(df, group_col, fill_cols):
    for col in fill_cols:
        df[col] = df.groupby(group_col)[col].transform(lambda x: x.fillna(x.mean()))
        
        nan_row = df[col].isna()
        
        if nan_row.any():
            df.loc[nan_row, col] = df[col].mean() 
            
    return df.head()

fill_nan(llm_df,'brand',['horsepower','volume'])
```

```python
#split into train and validation 
train_x, val_x, train_y, val_y = train_test_split(scaled_train, train_target, test_size = 0.2, random_state = 42)
```

```python
#define a set of models 

from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
import xgboost as xgb


def train_and_evaluate_models(train_x, val_x, train_y, val_y):
    models = {
        "Decision Tree": DecisionTreeRegressor(random_state=42),
        "Random Forest": RandomForestRegressor(n_estimators=50, random_state=42),
        "Gradient Boosting": GradientBoostingRegressor(n_estimators=50, learning_rate=0.1, random_state=42),
        "SVR": SVR(kernel='rbf', C=1.0, epsilon=0.1),
        "XGBoost": xgb.XGBRegressor(objective='reg:squarederror', max_depth=3, learning_rate=0.1, n_estimators=50)
    }

    results = {}

    for name, model in models.items():
        # Train the model
        model.fit(train_x, train_y)
        # Make predictions
        pred_y = model.predict(val_x)
        # Calculate RMSE
        rmse = np.sqrt(mean_squared_error(val_y, pred_y))
        # Store the results
        results[name] = rmse
        print(f"{name} RMSE: {rmse}")

    return results

#train and evaluate models
results = train_and_evaluate_models(train_x, val_x, train_y, val_y)

#print results
print("\nModel Performance:")
for model, rmse in results.items():
    print(f"{model}: RMSE = {rmse}")
```

# Perform hyperparameter tuning on it and increase the number of trees 

```python
from sklearn.model_selection import GridSearchCV
```

```python
#define parameter grid
param_grid = {
    'n_estimators': [100, 200, 500],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2]
}

xgb_model = xgb.XGBRegressor(objective='reg:squarederror', random_state=42)

grid_search = GridSearchCV(estimator=xgb_model, param_grid=param_grid, cv=3, 
                           scoring='neg_mean_squared_error', n_jobs=-1, verbose=1)

grid_search.fit(train_x, train_y)

#get the best parameters
best_params = grid_search.best_params_
print(f"Best Parameters: {best_params}")

#train the model with the best parameters
best_xgb = grid_search.best_estimator_
best_xgb.fit(train_x, train_y)

#make predictions and evaluate 

pred_y = best_xgb.predict(val_x)
rmse = np.sqrt(mean_squared_error(val_y, pred_y))
print(f"RMSE: {rmse}")


#plot the validation and predicted target values on the same figure
plt.figure(figsize=(14, 6))

plt.plot(val_y, color='blue', label='Validation')
plt.plot(pred_y, color='orange', label='Predicted')

plt.title('Line Plot of Train and Validation')
plt.xlabel('Index')
plt.ylabel('Price')
plt.legend()
```
