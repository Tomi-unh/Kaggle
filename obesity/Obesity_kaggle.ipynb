```python
#build the model 
#build a neural network model to get more used to pytorch

import torch
import torch.nn as nn
import torch.optim as optim
```

```python
#convert the data to pytorch tensors 

train_array = train_df.values
test_array = test_df.values

# Convert the NumPy arrays to PyTorch tensors
train_tensor = torch.tensor(train_array, dtype=torch.float)
test_tensor = torch.tensor(test_array, dtype=torch.float)
```

```python
train_tensor
```

```python
np.shape(train_tensor)
```

```python
target_array = target_dummies.values
target_tensor = torch.tensor(target_array, dtype = torch.float)
```

```python
import numpy as np
np.shape(test_tensor)[1]
```

```python
import torch.nn.functional as F
from sklearn.model_selection import train_test_split
class ObesityModel(nn.Module):
    def __init__(self, input_dim, hidden_size, classes):
        super(ObesityModel,self).__init__()
        
        self.input_dim = input_dim
        self.hidden_size = hidden_size
        self.classes = classes
        
        self.fc1 = nn.Linear(input_dim, hidden_size[0])  # Input layer.
        
        # Define additional hidden layers
        self.hidden_layers = nn.ModuleList()
        for i in range(len(hidden_size) - 1):
            self.hidden_layers.append(nn.Linear(hidden_size[i], hidden_size[i+1]))
        
        # Output layer
        self.output_layer = nn.Linear(hidden_size[-1], classes)
        
        #add softmax layer
        self.softmax = nn.Softmax(dim=1)
        
    def forward(self, x):
        out = F.relu(self.fc1(x))
        for hidden_layer in self.hidden_layers:
            out = F.relu(hidden_layer(out))
           
        #define the out layer
        out = self.output_layer(out)
        out = self.softmax(out)
        return out
    
    
    def predict(self, x):
        self.eval()
        with torch.no_grad():
            out = self.forward(x)
            return out
    
input_dim = np.shape(train_tensor)[1]
hidden_size = [128,64,32] #tuple for the number of dense layers
classes = np.shape(target_tensor)[1]

model = ObesityModel(input_dim,hidden_size,classes)

#split the data into training and testing splits
X_train, X_test, y_train, y_test = train_test_split(train_tensor, target_tensor, test_size=0.15, random_state=42)

#split the train data into training and validation splits    
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.1765, random_state=42)


criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

#parameters for early stopping
best_val_loss = float('inf')
patience = 10
counter = 0

# Training loop
num_epochs = 1000
for epoch in range(num_epochs):
    # Forward pass
    outputs = model(X_train)
    loss = criterion(outputs, y_train)
    
    # Backward pass and optimization
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    # Validation phase
    model.eval()
    with torch.no_grad():
        val_outputs = model(X_val)
        val_loss = criterion(val_outputs, y_val.argmax(dim=1))
    
    # Check for early stopping
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        counter = 0
    else:
        counter += 1
        if counter >= patience:
            print(f'Early stopping at epoch {epoch}')
            break
            
    # Print progress
    if (epoch + 1) % 10 == 0:
        print(f'Epoch [{epoch + 1}/{num_epochs}], Train Loss: {loss.item():.4f}, Val Loss: {val_loss.item():.4f}')
              
```

```python
from sklearn.metrics import roc_auc_score, accuracy_score

# calculate the ROC AUC score
y_pred = model.predict(X_test)
roc_auc = roc_auc_score(y_test, y_pred)
print("ROC AUC Score:", roc_auc)


predicted_ind = np.argmax(y_pred, axis=1)
actual_ind = np.argmax(y_test, axis = 1)

#get the accuracy on the test dataset
accuracy = accuracy_score(actual_ind,predicted_ind)
print("Accuracy:", accuracy)

```
